```python
import pathlib
import textwrap

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown

def to_markdown(text):
  text = text.replace('•', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
import requests
from bs4 import BeautifulSoup as bs

# "저칼로리" 키워드로 검색한 모바일 만개의레시피 페이지
url = "https://m.10000recipe.com/recipe/list.html?q=%EC%A0%80%EC%B9%BC%EB%A1%9C%EB%A6%AC"
response = requests.get(url)
soup = bs(response.text, "html.parser")

# 1) 레시피 목록을 감싸고 있는 <div class="common_sp_list_li"> 요소들 찾기
recipe_divs = soup.find_all("div", class_="common_sp_list_li")

for div in recipe_divs:
    # 2) 각 레시피 블록 안에서 <a class="common_sp_link"> 태그 찾기 (레시피 상세로 이동하는 링크)
    a_tag = div.find("a", class_="common_sp_link")
    if not a_tag:
        continue

    # 3) 레시피 제목 텍스트 추출
    title = a_tag.get_text(strip=True)
    print("레시피 제목:", title)
```

```python
#ChatSession 객체 반환
chat_session = model.start_chat(history=[])

user_queries = ["다이어트 하는 중인데 크림파스타의 칼로리는 몇인가요?", "칼로리가 높다면 대체 식품 및 레시피를 알려주세요."]
for user_query in user_queries:
    print(f'[사용자]: {user_query}')
    response = chat_session.send_message(user_query)
    print(f'[CC]: {response.text}') #CC는 챗봇 모델 이름!
```

```python
model = genai.GenerativeModel('gemini-1.5-flash')
user_queries = [{'role':'user', 'parts': ["크림파스타의 칼로리를 알려주세요."]},
                {'role':'user', 'parts': ["칼로리가 높다면 대체 식품 및 레시피를 알려주세요."]}
            ]
history = []
for user_query in user_queries:
    history.append(user_query)
    print(f'[사용자]: {user_query["parts"][0]}')
    response = model.generate_content(history)
    print(f'[CC]: {response.text}')
    history.append(response.candidates[0].content)
```

```python
system_instruction = "당신은 다이어트 및 식단 조절을 도와주는 칼로리 코치입니다. 사용자는 다이어트와 식단 조절을 원하는 사람입니다. 간단하게 사용자가 물어본 음식의 칼로리를 말해주고 저칼로리 대체 식품이나 레시피를 알려주면 됩니다. 3문장 이내로 짧게 얘기하세요."
model = genai.GenerativeModel("gemini-1.5-flash", system_instruction=system_instruction)

chat_session = model.start_chat(history=[])  # ChatSession 객체 반환
user_queries = ["다이어트 중인데 떡볶이 음식의 칼로리가 궁금해요.", "대체할 수 있는 저칼로리 음식 레시피 알려주세요.", "떡볶이를 먹고 싶은데 저칼로리로 먹을 수 있는 방법이 없나요?","레시피가 궁금해요."]

for user_query in user_queries:
    print(f"[사용자]: {user_query}")
    response = chat_session.send_message(user_query)
    print(f"[CC]: {response.text}")
```

```python
import streamlit as st
st.title("echo-bot")

if "chat_history" not in st.session_state:
    st.session_state.chat_history = []

for content in st.session_state.chat_history:
    with st.chat_message(content["role"]):
        st.write(content['message'])

if prompt := st.chat_input("메시지를 입력하세요."):
    with st.chat_message("user"):
        st.write(prompt)
        st.session_state.chat_history.append({"role": "user", "message": prompt})

    with st.chat_message("ai"):
        response = prompt
        st.write(response)
        st.session_state.chat_history.append({"role": "ai", "message": response})
```

```python
import google.generativeai as genai
import streamlit as st

# genai.configure(api_key=GOOGLE_API_KEY)
YOUR_API_KEY = 'YourKey'
genai.configure(api_key=YOUR_API_KEY)

st.title("Gemini-Bot")

@st.cache_resource
def load_model():
    model = genai.GenerativeModel('gemini-1.5-flash')
    print("model loaded...")
    return model

model = load_model()

if "chat_session" not in st.session_state:
    st.session_state["chat_session"] = model.start_chat(history=[])

for content in st.session_state.chat_session.history:
    with st.chat_message("ai" if content.role == "model" else "user"):
        st.markdown(content.parts[0].text)

if prompt := st.chat_input("메시지를 입력하세요."):
    with st.chat_message("user"):
        st.markdown(prompt)
    with st.chat_message("ai"):
        response = st.session_state.chat_session.send_message(prompt)
        st.markdown(response.text)
```
